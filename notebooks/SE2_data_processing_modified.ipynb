# SE2_data_processing


```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Paths 
raw_dir = Path.home()/"data"/"raw"
processed_dir = Path.home()/"data"/"processed"
processed_dir.mkdir(parents=True, exist_ok=True)

# Input file from SE1
raw_file = raw_dir/"MPs_WWTPs_Eff_DUMMY.csv"

# Load data
df = pd.read_csv(raw_file)

df.head()
```

```python
import numpy as np

# Standardize column names (remove spaces)
df.columns = df.columns.str.replace(" ", "_")

# Select concentration columns (11 to 52 exclude the last column total conc)
conc_cols = df.columns[10:-1]

# STEP 1: Remove rows where ALL concentration values are missing
df = df[df[conc_cols].notna().any(axis=1)]

# STEP 2: Remove rows with negative values in ANY compound
df = df[(df[conc_cols] >= 0).all(axis=1)]

# STEP 3: Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# STEP 4: Remove rows with invalid dates
df = df[df["Date"].notna()]

df.head()
```

```python
from pathlib import Path

# Define processed directory (same as SE2)
analysis_dir = Path.home() / "data" / "processed"
analysis_dir.mkdir(parents=True, exist_ok=True)

# Define output file
processed_file = analysis_dir/"analysis_ready_MPs_data.csv"

# Save dataframe
df.to_csv(processed_file, index=False)

processed_file
```
